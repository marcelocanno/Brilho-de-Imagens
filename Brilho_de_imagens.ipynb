# Aumentando o brilho de uma imagem com python e opencv

## importe as bibliotecas necessárias
- opencv - biblioteca de processamento de imagem
- numpy - necessária para criar a imagem de brilho
- cv2_imshow - para o google colab mostrar as imagens


# Install Library

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

## Leia a imagem

```python
#caminho da imagem
imagePath = '/content/Flor III.jpg'

#lendo a imagem
image = cv2.imread(imagePath)

cv2_imshow(image)
```

## Crie a matriz de brilho onde cada ponto será (100,100,100) e adicione a imagem original
Lembrando que o valor máximo do pixel no espaço RGB é 255.

```python
#criando image imagem do mesmo tamanho para ser o 'brilho'
bright = np.zeros_like(image)
#adicionaremos um brilho de 100 em cada pixel
bright[:,:,:] = [0,100,100] # (b,g,r)
#some 100 em cada pixel da imagem
brilho = cv2.add(image, bright)
#caso a soma ultrapasse 255 (máximo), limite-o a 255 (branco)
brilho[brilho > 255] = 255
```

## Mostre a imagem

```python
#mostre a imagem original
cv2_imshow(image)
```

```python
#mostre a imagem de brilho
cv2_imshow(brilho)
```

```python
#espere um comando
cv2.waitKey()
```
